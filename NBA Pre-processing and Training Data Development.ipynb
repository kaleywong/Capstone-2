```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Importing DataFrame
df = pd.read_csv('2022-2023_NBA_Player_Stats_Regular.csv', delimiter = ';', encoding = 'ISO-8859-1', index_col=0)
df.head(5)
```

```python
len(df)
```

```python
all_team_players = ['Giannis Antetokounmpo', 'LeBron James', 'Jayson Tatum', 'Donovan Mitchell', 'Kyrie Irving', 'Luka Don?i?',
                  'Nikola Joki?', 'Joel Embiid', 'Ja Morant', 'Lauri Markkanen', 'Bam Adebayo', 'Jaylen Brown', 'DeMar DeRozan',
                   'Tyrese Haliburton', 'Jrue Holiday', 'Julius Randle', 'Pascal Siakam', 'Anthony Edwards', "De'Aaron Fox", 'Paul George',
                   'Shai Gilgeous-Alexander', 'Jaren Jackson Jr.', 'Damian Lillard', 'Stephen Curry', 'Kevin Durant','Zion Williamson',
                   'Domantas Sabonis']

df['Made_All_Team'] = 0
df.loc[df['Player'].isin(all_team_players), 'Made_All_Team'] = 1
df[df['Made_All_Team'] == 1]
```

```python
important_columns = ['Player', 'PTS', 'AST', 'TRB', 'FG%', '3P%', 'BLK', 'STL', 'MP', 'FT%', 'Made_All_Team']
new_df = df[important_columns]
new_df.head()
```

 Since there are no categorical features in the new data frame, there is no need to create dummy variables

## Extracting Features and Target Variable
The features are the columns that represent different aspects of player performance (PTS, AST, TRB, FG%, 3P%, BLK, STL, MP, FT%)

The target variable is the column 'Made_All_Team', which indicates whether a player made an all-star team or not (1 if yes, 0 if no

```python
features = new_df.drop(['Player', 'Made_All_Team'], axis=1)
target = new_df['Made_All_Team']
```

## Standardize the Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

## Split the Data into Training and Testing Sets
We specify the test size as 0.2, which means that 20% of the data will be used for testing, and the remaining 80% will be used for training

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_features, target, test_size=0.2, random_state=42)
```

## Checking the shapes of the subsets

```python
print("Training features shape:", X_train.shape)
print("Testing features shape:", X_test.shape)
print("Training target shape:", y_train.shape)
print("Testing target shape:", y_test.shape)
```
