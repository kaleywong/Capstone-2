## Hypothesis Testing 
We want to see if there are significant differences in various performance metrics between NBA Players who made the All-NBA Team and those who did not.We will focus on the features I selected above and will run a two-sample t-test for each of the features to compare the means of the featre between the two groups (players who made All-NBA Team vs. those who did not).

## Hypothesis Forumation
##### Points per Game (PTS):

Null Hypothesis (H0): There is no significant difference in the average points per game between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average points per game between players who made the All-NBA Team and those who did not.

##### Assists per Game (AST):

Null Hypothesis (H0): There is no significant difference in the average assists per game between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average assists per game between players who made the All-NBA Team and those who did not.

##### Total Rebounds per Game (TRB):

Null Hypothesis (H0): There is no significant difference in the average total rebounds per game between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average total rebounds per game between players who made the All-NBA Team and those who did not.

##### Field Goal Percentage (FG%):

Null Hypothesis (H0): There is no significant difference in the average field goal percentage between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average field goal percentage between players who made the All-NBA Team and those who did not.

##### 3-Point Field Goal Percentage (3P%):

Null Hypothesis (H0): There is no significant difference in the average 3-point field goal percentage between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average 3-point field goal percentage between players who made the All-NBA Team and those who did not.

##### Blocks per Game (BLK):

Null Hypothesis (H0): There is no significant difference in the average blocks per game between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average blocks per game between players who made the All-NBA Team and those who did not.

##### Steals per Game (STL):

Null Hypothesis (H0): There is no significant difference in the average steals per game between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average steals per game between players who made the All-NBA Team and those who did not.

##### Minutes Played per Game (MP):

Null Hypothesis (H0): There is no significant difference in the average minutes played per game between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average minutes played per game between players who made the All-NBA Team and those who did not.

##### Free Throw Percentage (FT%):

Null Hypothesis (H0): There is no significant difference in the average free throw percentage between players who made the All-NBA Team and those who did not.

Alternative Hypothesis (H1): There is a significant difference in the average free throw percentage between players who made the All-NBA Team and those who did not.


## Hypothesis Testing

```python
from scipy.stats import ttest_ind

features = ['PTS', 'AST', 'TRB', 'FG%', '3P%', 'BLK', 'STL', 'MP', 'FT%']


for feature in features:
    # Extract data for players who made All-NBA Team and those who didn't
    all_nba_data = new_df[new_df['Made_All_Team'] == 1][feature]
    not_all_nba_data = new_df[new_df['Made_All_Team'] == 0][feature]

    # Perform two-sample t-test
    t_statistic, p_value = ttest_ind(all_nba_data, not_all_nba_data)

    # Interpret the results
    alpha = 0.05
    print(f"Feature: {feature}")
    print("T-statistic:", t_statistic)
    print("P-value:", p_value)

    if p_value < alpha:
        print(f"Reject the null hypothesis. There is a significant difference in the average {feature} between players who made the All-NBA Team and those who did not.")
    else:
        print(f"Fail to reject the null hypothesis. There is no significant difference in the average {feature} between players who made the All-NBA Team and those who did not.")
    print()
```

## Key Findings
##### Significant Differences
- We found significant differences in several performance metrics between players who made the All-NBA Team and those who did not. Specifically,  points per game (PTS), assists per game (AST), total rebounds per game (TRB), blocks per game (BLK), steals per game (STL), minutes played per game (MP), and free throw percentage (FT%) showed statistically significant disparities between the two groups.

##### Non-Significant Differences
- Notably, field goal percentage (FG%) and 3-point field goal percentage (3P%) did not exhibit significant differences between players who made the All-NBA Team and those who did not.

### Implications
The findings highlight the importance of certain performance metrics, such as points, assists, rebounds, blocks, steals, minutes played, and free throw percentage, in determining All-NBA Team selection.
Players with exceptional performances in these areas are more likely to be chosen for the All-NBA Team, as evidenced by the statistically significant differences observed.
